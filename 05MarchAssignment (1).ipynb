Q1. How can you create a Bokeh plot using Python code?

Bokeh is a Python library used for creating interactive visualizations for modern web browsers. 

First, we need to install the Bokeh library using pip command:

pip install bokeh

After that, import the necessary functions and classes from the Bokeh library

from bokeh.plotting import figure ,output_file , show



```python
from bokeh.plotting import figure, output_file, show

x_values = [1, 2, 3, 4, 5]
y_values = [5, 4, 3, 2, 1]

p = figure(title='Bokeh Plot', x_axis_label='X Axis', y_axis_label='Y Axis')

p.circle(x=x_values, y=y_values, size=10, color='blue')

output_file('bokeh_plot.html')

show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

In Bokeh, glyphs are visual marks or symbols that are used to represent data in a plot. They can be used to create scatter plots, line plots, bar charts, and other types of visualizations. Bokeh provides a wide range of glyphs, including circles, squares, triangles, lines, and more.


To add glyphs to a Bokeh plot, you can create a glyph renderer and add it to the plot using the add_glyph() method of the plot. The add_glyph() method takes two arguments: the data source and the glyph specification.

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource


x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
colors = ['red', 'green', 'blue', 'orange', 'purple']


source = ColumnDataSource(data=dict(x=x, y=y, colors=colors))


p = figure(title="My Plot", x_axis_label='X', y_axis_label='Y')
p.circle('x', 'y', size=10, color='colors', source=source)


p.square(x=[2.5, 3.5], y=[5, 5], size=20, color='black')


show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

Bokeh server is a component of Bokeh that provides a way to create interactive plots and applications that can be updated in real-time, without the need for manual intervention.

Bokeh server works by allowing the user to define a Python function or class that generates a plot or application. 



To create an interactive plot using Bokeh server, you can follow these general steps:

1. Define your data and create a plot using Bokeh's plotting API.

2. Define a Python function or class that modifies the plot in response to user input.

3. This function or class should take a doc argument, which represents the document being served by the Bokeh server.

4. Create a Bokeh server application using the bokeh.application module, and add your function or class to the application.

5. Run the Bokeh server using the bokeh.server module.
